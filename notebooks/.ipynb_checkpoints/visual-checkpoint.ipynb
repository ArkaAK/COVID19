```python
def diff(x):
    t= list(x.sort_values(by="Date Announced",ascending=True)["Patient Number"].diff())[-1]
    return t
x= df.groupby(["States","Latitude","Longitude","Date Announced"],as_index=False)["Patient Number"].count()
x.groupby(["States","Latitude","Longitude"],as_index=False).apply(diff)
```
